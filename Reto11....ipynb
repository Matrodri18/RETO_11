##Procesar el archivo y extraer:

###- Cantidad de vocales
###- Cantidad de consonantes
###- Listado de las 50 palabras que más se repiten

```python
import requests

def cont_vocales(texto): # Funcion para contar las vocales
    vocales = 'aeiouÁÉÍÓÚáéíóú'
    cont = {vocal: texto.lower().count(vocal) for vocal in vocales}
    return cont

def cont_consonantes(texto): # funcion para contar las consonantes
    consonantes = 'bcdfghjklmnñpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ'
    cont = sum(texto.lower().count(n) for n in consonantes)
    return cont

def palabras_repetidas(texto, n=50):
    # Convertir a minúsculas para tomar todas las palabras iguales y dividir en palabras
    palabras = texto.lower().split()

    # Creamos un diccionario que contabilice la frecuencia de las palabras
    frecuencias = {}
    for palabra in palabras:
      frecuencias[palabra] = frecuencias.get(palabra, 0) + 1

    # se ordenan por frecuencia de mayor a menor y luego alfabeticamente
    mas_frecuentes = sorted(frecuencias.items(), key=lambda x: (-x[1], x[0]))[:n]
    return mas_frecuentes

def main():
    enlace = 'https://www.py4e.com/code3/mbox.txt'  # ingresamos al archivo
    response = requests.get(enlace)
    texto = response.text  # lo convertimos en una cadena o a string

    if texto:
        # Contamos las vocales
        conteo_vocales = cont_vocales(texto)

        # se realiza el print del conteo correspondiente a las
        print("Conteo de vocales:")
        for vocal, cantidad in conteo_vocales.items():
            print(str(vocal) + " : " + str(cantidad))
        print("Total de vocales: " + str(sum(conteo_vocales.values()))) # se imprime el numero total de vocales
        print("Total de consonantes: " + str(cont_consonantes(texto))) # se imprime el numero total de consonantes
        # Imprimimos el listado de las 50 palabras mas comunes del archivo
        print("Las palabras mas comunes son....")
        for palabra, frecuencia in palabras_repetidas(texto):
            print(str(palabra) + " : " + str(frecuencia))
    else:
        print("No abre el achivo") # print en caso tal de algun fallo en el archivo

if __name__ == "__main__":
    main()
```
